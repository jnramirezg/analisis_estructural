# **Componentes de la deformación en un punto**

El campo vectorial de desplazamientos en 2D es:

$\underline{u}(x,y)=[u(x,y), v(x,y)]^T$

En general, si tenemos una posición inicial $\underline{P}=(P_x, P_y)$, después del desplazamiento, tenemos una posición final $\underline{P'}$:

$\underline{P'}= \underline{P} + \underline{u}(\underline{P})=[P_x+u(\underline{P}),P_y+v(\underline{P})]^T$

Por simplicidad en el análisis estudiamos el caso 2D. Considere un sólido como el mostrado en la figura, que se encuentra bajo la acción de fuerzas superficiales y másicas. Y que de acuerdo con el modelo de esfuerzos, al interior del sólido, todo punto $(x, y)$ está sometido a los esfuerzos normales $\sigma_x(x, y)$ y $\sigma_y(x, y)$ y a un esfuerzo tangencial $\tau_{xy}(x, y)$. Estos esfuerzos producirán desplazamientos en las direcciones de los ejes $x$ y $y$: $u(x, y)$ y $v(x, y)$
para todo punto $(x, y)$ perteneciente al sólido y harán, además, que el sólido $ABCD$ se deforme hasta alcanzar la posición $A′B′C′D′$. Supondremos adicionalmente que los desplazamientos $u$ y $v$ son funciones continuas con derivadas continuas (por la hipótesis de continuidad de todo sólido elástico).

Si queremos analizar la deformación del sólido en el punto $A$, debemos estudiar qué pasa en el sólido $ABCD$ y, luego, aplicar límites de modo tal que $ABCD$ tienda al punto A.

Determinamos las coordenadas de los puntos $\underline{A}$, $\underline{B}$, $\underline{C}$ y $\underline{D}$:

$\underline{A}=[A_x, A_y]^T=\color{red}{[x, y]^T}$\
$\underline{B}=[B_x, B_y]^T=\color{red}{[x+\Delta x, y]^T}$\
$\underline{C}=[C_x, C_y]^T=\color{red}{[x+\Delta x, y+\Delta y]^T}$\
$\underline{D}=[D_x, D_y]^T=\color{red}{[x, y+\Delta y]^T}$


Determinamos las coordenadas de los puntos $\underline{A'}$, $\underline{B'}$, $\underline{C'}$ y $\underline{D'}$:

$\underline{A'}=\underline{A} + \underline{u}(\underline{A})=[A'_x, A'_y]^T=\color{red}{[A_x+u(\underline{A}), A_y+v(\underline{A})]^T}=[x + u(x,y), y + v(x,y)]^T$

$\underline{B'}=\underline{B} + \underline{u}(\underline{B})=[B'_x, B'_y]^T=\color{red}{[B_x+u(\underline{B}), B_y+v(\underline{B})]^T}=[x+\Delta x + u(x+\Delta x,y), y + v(x+\Delta x,y)]^T$

$\underline{C'}=\underline{C} + \underline{u}(\underline{C})=[C'_x, C'_y]^T=\color{red}{[C_x+u(\underline{C}), C_y+v(\underline{C})]^T}=[x+\Delta x + u(x+\Delta x,y+\Delta y), y+\Delta y + v(x+\Delta x,y+\Delta y)]^T$

$\underline{D'}=\underline{D} + \underline{u}(\underline{D})=[D'_x, D'_y]^T=\color{red}{[D_x+u(\underline{D}), D_y+v(\underline{D})]^T}=[x + u(x,y+\Delta y), y+\Delta y + v(x,y+\Delta y)]^T$

Determinamos las coordenadas de los puntos $\underline{B_2}$ y $\underline{D_2}$:

$B_2=\color{red}{[B'_x, A'_y]^T}$

$D_2=\color{red}{[A'_x, D'_y]^T}$

Se importan algunas funciones de la librería Sympy para cálculos simbólicos:

```python
from sympy import symbols, Lambda, Function, Point
```

Se crean las variables simbólicas $x$, $y$, $\Delta x$ y $\Delta y$ que son suficientes para describir las coordenadas de los puntos:

```python
x,   y = symbols("x y")
Dx, Dy = symbols("Delta_x Delta_y")
```

Se crean las funciones de las componentes del campo vectorial de desplazamientos.`Function` garantiza que se inteprete como una función matemática y `Lambda`permite que sea una función desde el punto de vista de la programación.

```python
u = Lambda((x,y), Function('u')(x,y))
v = Lambda((x,y), Function('v')(x,y))
```

Definimos las coordenadas de los puntos $\underline{A}$, $\underline{B}$, $\underline{C}$ y $\underline{D}$:

$\underline{A}=\color{red}{[x, y]^T}$\
$\underline{B}=\color{red}{[x+\Delta x, y]^T}$\
$\underline{C}=\color{red}{[x+\Delta x, y+\Delta y]^T}$\
$\underline{D}=\color{red}{[x, y+\Delta y]^T}$


```python
A = Point(x   , y   )
B = Point(x+Dx, y   )
C = Point(x+Dx, y+Dy)
D = Point(x   , y+Dy)
```

Definimos las coordenadas de los puntos $\underline{A'}$, $\underline{B'}$, $\underline{C'}$ y $\underline{D'}$:

$\underline{A'}=\color{red}{[A_x+u(\underline{A}), A_y+v(\underline{A})]^T}$

$\underline{B'}=\color{red}{[B_x+u(\underline{B}), B_y+v(\underline{B})]^T}$

$\underline{C'}=\color{red}{[C_x+u(\underline{C}), C_y+v(\underline{C})]^T}$

$\underline{D'}=\color{red}{[D_x+u(\underline{D}), D_y+v(\underline{D})]^T}$

```python
Ap = Point(A.x + u(*A), A.y + v(*A))
Bp = Point(B.x + u(*B), B.y + v(*B))
Cp = Point(C.x + u(*C), C.y + v(*C))
Dp = Point(D.x + u(*D), D.y + v(*D))
```

Definimos las coordenadas de los puntos $\underline{B_2}$ y $\underline{D_2}$:

$B_2=\color{red}{[B'_x, A'_y]^T}$

$D_2=\color{red}{[A'_x, D'_y]^T}$

```python
B2 = Point(Bp.x, Ap.y)
D2 = Point(Ap.x, Dp.y)
```

Se calculan las distancias $A'B_2$, $AB$, $A'D_2$, $AD$, $D_2D'$ y $B_2B'$ requeridas en las ecuaciones:

```python
ApB2 = B2.distance(Ap)
AB   =  B.distance(A)
ApD2 = D2.distance(Ap)
AD   =  D.distance(A)
D2Dp = Dp.distance(D2)
B2Bp = Bp.distance(B2)
```

## (1) Cálculo de $\varepsilon_x$:
Teniendo en cuenta que: $$\varepsilon_x = \lim_{\substack{ \Delta_x \to 0\\ \Delta_y \to 0}}\dfrac{A'B_2-AB}{AB}$$

```python
(ApB2 - AB)/AB
```

Sacando de las raíces y aplicando el límite:
$$\require{cancel}$$
$$\varepsilon_x = \lim_{\substack{ \Delta_x \to 0\\ \Delta_y \to 0}}\dfrac{\cancel{-\Delta x} + \cancel{\Delta x} - u(x,y) + u(x+\Delta x,y)}{\Delta x} = \lim_{\Delta_x \to 0}\dfrac{ u(x+\Delta x,y) - u(x,y)}{\Delta x}$$

Es decir, por la definición de la derivada parcial respecto a $x$ de la función $u(x,y)$: $$\require{boxed}$$ $$\boxed{\varepsilon_x=\dfrac{\partial u}{\partial x}}$$


## (2) Cálculo de $\varepsilon_y$:
Teniendo en cuenta que: $$\varepsilon_y = \lim_{\substack{ \Delta_x \to 0\\ \Delta_y \to 0}}\dfrac{A'D_2-AD}{AD}$$

```python
(ApD2 - AD)/AD
```

Sacando de las raíces y aplicando el límite:
$$\require{cancel}$$
$$\varepsilon_y = \lim_{\substack{ \Delta_x \to 0\\ \Delta_y \to 0}}\dfrac{\cancel{-\Delta y} + \cancel{\Delta y} - v(x,y) + v(x,y+\Delta y)}{\Delta y} = \lim_{\Delta_y \to 0}\dfrac{ v(x,y+\Delta y) - v(x,y)}{\Delta y}$$

Es decir, por la definición de la derivada parcial respecto a $y$ de la función $u(x,y)$: $$\require{boxed}$$ $$\boxed{\varepsilon_y=\dfrac{\partial v}{\partial y}}$$


## (3) Cálculo de $\gamma_{xy}$:
Teniendo en cuenta que: $$\gamma_{xy} = \lim_{\substack{ \Delta_x \to 0\\ \Delta_y \to 0}}\left(\dfrac{B_2B'}{A'B_2}+\dfrac{D_2D'}{A'D_2}\right)$$

```python
B2Bp/ApB2 + D2Dp/ApD2
```

Sacando de las raíces y aplicando el límite:
$$\require{cancel}$$
$$\gamma_{xy} = \lim_{\substack{ \Delta_x \to 0\\ \Delta_y \to 0}}\left(\dfrac{u(x,y + \Delta_y)-u(x,y)}{\Delta y + v(x,y+\Delta y)-v(x,y)}+\dfrac{v(x+\Delta x,y)-v(x,y)}{\Delta x + u(x+\Delta x,y)-u(x,y)}\right)$$

Multiplicando el primer término por $1=\dfrac{\frac{1}{\Delta y}}{\frac{1}{\Delta y}}$ y el segundo término por $1=\dfrac{\frac{1}{\Delta x}}{\frac{1}{\Delta x}}$
$$\gamma_{xy} = \lim_{\Delta_y \to 0}\dfrac{\dfrac{u(x,y + \Delta_y)-u(x,y)}{\Delta y}}{\dfrac{\Delta y + v(x,y+\Delta y)-v(x,y)}{\Delta y}}+\lim_{\Delta_x \to 0}\dfrac{\dfrac{v(x+\Delta x,y)-v(x,y)}{\Delta x}}{\dfrac{\Delta x + u(x+\Delta x,y)-u(x,y)}{\Delta x}}$$

Simplificando y aplicando el concepto de derivada parcial:
$$\gamma_{xy}=\dfrac{\frac{\partial u}{\partial y}}{1+\frac{\partial v}{\partial y}}+\dfrac{\frac{\partial v}{\partial x}}{1+\frac{\partial u}{\partial x}}$$

Pero $\varepsilon_x=\frac{\partial u}{\partial x}$ y $\varepsilon_y=\frac{\partial v}{\partial y}$, entonces: $$\gamma_{xy}=\dfrac{\frac{\partial u}{\partial y}}{1+\varepsilon_y}+\dfrac{\frac{\partial v}{\partial x}}{1+\varepsilon_x}$$

Pero gracias a la H.5. (cuerpo rígido - deformaciones pequeñas), $\varepsilon_x<<0$ y $\varepsilon_y<<0$, por lo tanto, $1+\varepsilon_x ≈ 1$ y $1+\varepsilon_y ≈ 1$. Es decir, $$\require{boxed}$$ $$\boxed{\gamma_{xy}=\frac{\partial u}{\partial y}+\frac{\partial v}{\partial x}}$$
